```python
import pandas as pd
```

```python
data=pd.read_csv('마켓컬리_레시피_라벨링용.csv')
```

```python
data
```

```python
title=pd.DataFrame(data['name'])
```

```python
title=title.drop_duplicates(['name'],keep='first').reset_index(drop=True)
```

```python
title.index
```

```python
title['code']=title.index
```

```python
title
```

```python
title=title[['code','name']]
```

```python
title.to_csv('food_title.csv')
```

```python
title
```

```python
data=pd.merge(data,title, how='left', on=None)
```

```python
data
```

```python
data=data[['id','code','name','ingredient','quantity']]
```

```python
data
```

```python
data.to_csv('food.csv')
```
